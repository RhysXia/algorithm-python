```python
"""
算法描述：
    有N件物品和一个承重为C的背包，每个物品的重量为Wi，价值为Pi
    求解将哪几件物品放入背包可以使总价值最大
"""


class Obj:
    """
    定义的物品类
    """

    def __init__(self, weight, price):
        """

        :param weight: 物品重量
        :param price:  物品价格
        :param status: 物品状态 0：未选中，1：被选中，2：不可选
        """
        self.weight = weight
        self.price = price
        self.status = 0
        pass


class Solution:
    """
    算法类
    """

    def __init__(self, objs=[], weight=0):
        """
        obj_list: 符合条件的物品列表
        can_weight: 背包还能承重大小
        can_choose_objs: 可选择的物品
        """
        self.choose_list = []
        self.can_weight = weight
        self.can_choose_objs = objs
        pass

    def getFitObj(self):
        _flag = -1
        _price = 0
        for index in range(len(self.can_choose_objs)):
            if self.can_choose_objs[index].status == 0:
                price = self.can_choose_objs[index].price / self.can_choose_objs[index].weight
                if price > _price:
                    _price = price
                    _flag = index
        return _flag
        pass

    def solve(self):
        while True:
            index = self.getFitObj()
            if index == -1:
                break
            obj = self.can_choose_objs[index]
            if obj.weight < self.can_weight:
                obj.status = 1
                self.choose_list.append(obj)
                self.can_weight -= obj.weight
            else:
                obj.status = 2
        return self.choose_list
        pass


def main():
    list = [Obj(35, 10), Obj(30, 40), Obj(60, 30), Obj(50, 50), Obj(40, 35), Obj(10, 40), Obj(25, 30)]
    s = Solution(list, 150)
    result = s.solve()
    for obj in result:
        print(obj.weight,obj.price)
    pass


if __name__ == '__main__':
    main()
```
